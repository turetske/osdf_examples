## Jetstream 2 and Exosphere

- Jetstream2 is a user-friendly cloud computing platform supported by the National Science Foundation (NSF). Before attempting to run any Jetstream related notebooks, please review their [getting started tutorial](https://docs.jetstream-cloud.org/getting-started/overview/)
- We will use their beginner friendly interface [Exosphere](https://docs.jetstream-cloud.org/getting-started/login/#click-add-allocation)
- Once you are able to get a Jetstream allocation and log in, please create at least an m3.medium instance with 8 CPUS and 30GB memory to run the notebooks. You can find more details about that [here](https://docs.jetstream-cloud.org/getting-started/first-instance/#give-your-instance-a-name). We recommend that you enable the 'Web Desktop' option in your cloud instance.

:::{note}: Jupyter Lab on Exosphere
Please refer to this [guide](https://docs.jetstream-cloud.org/general/jupyter/?h=jupyter) to spin up a jupyter lab session in your session. You can create a conda environment (using the requirements.txt in this repository) in your exosphere instance and use the same kernel to run the jupyter notebooks
:::
